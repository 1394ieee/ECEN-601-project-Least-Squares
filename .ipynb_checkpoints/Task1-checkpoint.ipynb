Task 1

Select the coefficients of Î² using the least squares approximation with data sets of size 11, 101, 1001, 10001, with equally spaced sample. Assess and plot the performance of each model using the MSE of (4).

```python
import pandas as pd
import math as math
import numpy as np
import matplotlib.pyplot as plt
#plot sin(x) in [0,10]
#for i in range(10,10000,10)
X=[]
Y=[]
n=1
for i in [10, 100, 1000, 10000]:
    l=np.arange(0,100001,i)
    x=l/10000
    y=np.sin(x)
    X.append(x)
    Y.append(y)
    size=np.array(X).size
xn=np.array(X)
yn=np.array(Y)

# show the original figure
fig, axs = plt.subplots(2, 2,figsize=(8,6))
fig.tight_layout(pad=3.0)
axs[0, 0].plot(xn[0], yn[0])
axs[0, 0].set_title('sin(x): data size= 10001')
axs[0, 1].plot(xn[1], yn[1], 'tab:blue')
axs[0, 1].set_title('sin(x): data size= 1001')
axs[1, 0].plot(xn[2],yn[2], 'tab:blue')
axs[1, 0].set_title('sin(x): data size= 101')
axs[1, 1].plot(xn[3], yn[3], 'tab:blue')
axs[1, 1].set_title('sin(x): data size= 11')
for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')
fig.savefig('T1_1.png')
```

```python
# find A and Î² for 0 set
A=[]
for P in range(0,11,1):
    B = np.power(xn[0],P)
    A.append(B) 
A0 = np.array(A)
A0=A0.T
#print(A)
Î²=np.dot(A0.T,A0);
Î²=np.linalg.inv(Î²);
Î²=np.dot(Î²,A0.T);
Î²0=np.dot(Î²,yn[0])

A1=[]
# find A and Î² for 1 set
for P in range(0,11,1):
    B = np.power(xn[1],P)
    A1.append(B) 
A1 = np.array(A1)
A1=A1.T
#print(A)
Î²=np.dot(A1.T,A1);
Î²=np.linalg.inv(Î²);
Î²=np.dot(Î²,A1.T);
Î²1=np.dot(Î²,yn[1])
#print (Î²)

A2=[]
# find A and Î² for 2 set
for P in range(0,11,1):
    B = np.power(xn[2],P)
    A2.append(B) 
A2 = np.array(A2)
A2=A2.T
#print(A)
Î²=np.dot(A2.T,A2);
Î²=np.linalg.inv(Î²);
Î²=np.dot(Î²,A2.T);
Î²2=np.dot(Î²,yn[2])

A3=[]
# find A and Î² for 3 set
for P in range(0,11,1):
    B = np.power(xn[3],P)
    A3.append(B) 
A3 = np.array(A3)
A3=A3.T
#print(A)
Î²=np.dot(A3.T,A3);
Î²=np.linalg.inv(Î²);
Î²=np.dot(Î²,A3.T);
Î²3=np.dot(Î²,yn[3])

#calculate MSE
ya0=np.dot(A0,Î²0)
MSE0=sum((yn[0]-ya0)**2)
ya1=np.dot(A1,Î²1)
MSE1=sum((yn[1]-ya1)**2)
ya2=np.dot(A2,Î²2)
MSE2=sum((yn[2]-ya2)**2)
ya3=np.dot(A3,Î²3)
MSE3=sum((yn[3]-ya3)**2)

#plot differences

fig, axs = plt.subplots(2, 2,figsize=(8,6))
fig.tight_layout(pad=4)
fig.subplots_adjust(top=0.88)

axs[0, 0].plot(xn[0], (yn[0]-ya0)**2)
axs[0, 0].set_title('data size= 10001,MSE =%.4f' %float(MSE0),pad=15)

axs[0, 1].plot(xn[1], (yn[1]-ya1)**2)
axs[0, 1].set_title('data size= 1001,MSE =%.4f' %float(MSE1),pad=15)
axs[1, 0].plot(xn[2], (yn[2]-ya2)**2)
axs[1, 0].set_title('data size= 101,MSE =%.4f' %float(MSE2),pad=15)
axs[1, 1].plot(xn[3], (yn[3]-ya3)**2)
axs[1, 1].set_title('data size= 11,MSE =%.4f' %float(MSE3),pad=15)
for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')
fig.suptitle('differences^2 at each location') 
fig.savefig('T1_2.png')
```
