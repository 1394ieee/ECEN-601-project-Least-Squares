Task 2
Define a new model by truncating the Taylorâ€™s series expansion of sin(x) to P = 10. Find the coefficients, and then assess the performance of this model.

```python
import pandas as pd
import math as math
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
from sympy import series, Symbol
from sympy.functions import sin, cos, exp
from sympy.plotting import plot


# Define symbol
x = Symbol('x')
# Function for Taylor Series Expansion
def taylor(function, x0, n):
    """
    Parameter "function" is our function which we want to approximate
    "x0" is the point where to approximate
    "n" is the order of approximation
    """
    return function.series(x,x0,n).removeO()
print('sin(x) =', taylor(sin(x), 0, 10))

# This will plot sine and its Taylor approximations
p = plot(sin(x),taylor(sin(x),0,10),(x,0,6),legend=True, show=False)
p[0].line_color = 'blue'
p[1].line_color = 'green'

p.title = 'Taylor Series Expansion for Sine'
p.show()

p.save('T2_1.png')
```

```python
#here we compare the results of sin(X) in range[0,5]
#AS when we only expand the P to 10, the approximate results would be far off the sin(X) results, 
#to get more similar results, higher p-degree is needed

l=np.arange(0,50001,10)
x0=l/10000
Y0=np.sin(x0)
y0 = x0**9/362880 - x0**7/5040 + x0**5/120 - x0**3/6 + x0
MSE0=sum((Y0-y0)**2)

l=np.arange(0,50001,100)
x1=l/10000
Y1=np.sin(x1)
y1 = x1**9/362880 - x1**7/5040 + x1**5/120 - x1**3/6 + x1
MSE1=sum((Y1-y1)**2)

l=np.arange(0,50001,1000)
x2=l/10000
Y2=np.sin(x2)
y2 = x2**9/362880 - x2**7/5040 + x2**5/120 - x2**3/6 + x2
MSE2=sum((Y2-y2)**2)

l=np.arange(0,50001,10000)
x3=l/10000
Y3=np.sin(x3)
y3 = x3**9/362880 - x3**7/5040 + x3**5/120 - x3**3/6 + x3
MSE3=sum((Y3-y3)**2)


#plot differences

fig, axs = plt.subplots(2, 2,figsize=(8,6))
fig.tight_layout(pad=4)
fig.subplots_adjust(top=0.88)

axs[0, 0].plot(x0, (Y0-y0)**2)
axs[0, 0].set_title('data size= 50001,MSE =%.3f' %float(MSE0),pad=15)

axs[0, 1].plot(x1, (Y1-y1)**2)
axs[0, 1].set_title('data size= 5001,MSE =%.3f' %float(MSE1),pad=15)
axs[1, 0].plot(x2, (Y2-y2)**2)
axs[1, 0].set_title('data size= 501,MSE =%.3f' %float(MSE2),pad=15)
axs[1, 1].plot(x3, (Y3-y3)**2)
axs[1, 1].set_title('data size= 51,MSE =%.3f' %float(MSE3),pad=15)
for ax in axs.flat:
    ax.set(xlabel='x', ylabel='y')
fig.suptitle('differences^2 at each location') 
fig.savefig('T2_2.png')
```
